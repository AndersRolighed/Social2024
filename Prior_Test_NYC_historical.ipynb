```python
df = pd.read_csv("NYC_Permitted_Event_Information_-_Historical_20240412.csv")
```

```python
print("Length of data: ", len(df))
for column in list(df.columns):
    print("Length of uniques in", column, " ", len(df[column].unique()))
```
